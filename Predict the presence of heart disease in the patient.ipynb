# Installing Independencies

```python
# Here we Gonna install Essintial libraries to get Started with our model
import numpy as numpy
import pandas as pandas
from sklearn.model_selection import train_test_split
from ucimlrepo import fetch_ucirepo
```

# Loading Data  and Displaying Statistics ðŸ“Š & Missing Values

```python
def loading_Data():

  heart_disease = fetch_ucirepo(id=45)

  global X_data_features , y_data_Target

  X_data_features = heart_disease.data.features

  y_data_Target = heart_disease.data.targets

  return X_data_features , y_data_Target

X_data_features , y_data_Target = loading_Data()



print('Name of Columns:' ,X_data_features.columns )

print('Numbs of the rows:' ,X_data_features.shape[0] )

print('Numbs of the Columns:' ,X_data_features.shape[1] )

# Separator : Used to Seperate Result to have better visuals , While Reviewing Output
print("-"*1000)

print('Display first 10 Rows of Data-set :' , "\n", X_data_features.head(10))


print('Display Last 10 Rows of Data-set :' , "\n", X_data_features.tail(10))


# Separator
print("-"*1000)

def Show_Basics_Statistics_DataSet():

  print("Show Fundemental (Basics) Statistics : " , "\n", X_data_features.describe())

  # Separator
  print("-"*1000)

  print("Show Missing Values : " , "\n", X_data_features.isnull().sum())


Show_Basics_Statistics_DataSet()
```

# Data - Preprocessing

```python
# Here we Will import necessary Modules to get Start in Data - preprocssing

from sklearn.impute import SimpleImputer # Here We import this module to handle and fix the missing Values

from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split # Here We import this module to Split the Data into => training and Testing

def Handle_Missing_ValuesAnd_Spiliting_Data_Into_Training_And_Testing():

  # Here we use mean strategy to fill Missing Values
  Data_Mean_imputation = SimpleImputer (strategy= 'mean')

  X_data_fitting = Data_Mean_imputation.fit_transform (X_data_features)

  global X_Data_For_Train , X_Data_For_Test , Y_Data_For_Train , Y_Data_For_Test # here We Make Varaibles Global to Access it Freely with Other Function

  X_Data_For_Train , X_Data_For_Test , Y_Data_For_Train , Y_Data_For_Test = train_test_split (X_data_fitting ,y_data_Target , test_size=0.2 , random_state=42 )

  # Here we use this line of code to check data that we split it
  print('Displaying Shapes of Features: ' , X_data_features.shape , X_Data_For_Train.shape , X_Data_For_Test.shape)

Handle_Missing_ValuesAnd_Spiliting_Data_Into_Training_And_Testing()


```

# Building and Evaluting Machine Learning Model


```python
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score , classification_report ,confusion_matrix

# Here we Make a function to Buid Support Vetor Machine Model
def Implement_Model (X_Data_For_Train , Y_Data_For_Train):

  SupportVector_Machine_Model = SVC(kernel='linear' , random_state=42)

  # Here we use this Line of Code to fit SVC Model to The Traning data
  SupportVector_Machine_Model.fit(X_Data_For_Train , Y_Data_For_Train )

  return SupportVector_Machine_Model

SupportVector_Machine_Model = Implement_Model(X_Data_For_Train, Y_Data_For_Train)

# Here we Make a function to Evalute the Preformance of our model That we build
def Model_Evalution_Preformance (SupportVector_Machine_Model , X_Data_For_Test , Y_Data_For_Test ):

  SVM__Prediction = SupportVector_Machine_Model.predict(X_Data_For_Test) # Here we make prediction on test - data by our model that we trained

  Svm_Accuracy_Score = accuracy_score( Y_Data_For_Test , SVM__Prediction ) # Computin / Calculating the accuracy Score of our model

  print('Accuracy for Suport Vector Machine Model (Svm)' , '\n' , Svm_Accuracy_Score )

  # Separator
  print("-"*1000)

  print('Classification Report for Suport Vector Machine Model (Svm)' , '\n' , classification_report(Y_Data_For_Test , SVM__Prediction))

  # Separator
  print("-"*1000)

  print('Confusion Matrix for Suport Vector Machine Model (Svm)' , '\n' , confusion_matrix(Y_Data_For_Test , SVM__Prediction))

Model_Evalution_Preformance (SupportVector_Machine_Model , X_Data_For_Test , Y_Data_For_Test )
```

# Fine tuning Machine learning Model

```python
from sklearn.model_selection import GridSearchCV

def Fine_Tuning_Model_By_Grid_SearchCV(model, X_Data_For_Train, Y_Data_For_Train):
    # Here as we can see we will Define the hyperparameters
    search_grid_params = {
        'C': [20, 40, 60],
        'gamma': [1, 0.1, 0.01, 0.001],
        'kernel': ['linear', 'rbf']
    }

    grid_search_cv = GridSearchCV(estimator=model, param_grid=search_grid_params, cv=5, scoring='accuracy', n_jobs=-1)

    #  Here we will try to fit Grid to Look out for best possible Hyperparameter by using Training Data
    grid_search_cv.fit(X_Data_For_Train, Y_Data_For_Train)

    # Print the best hyperparameters found
    print("Hyper-Parameter Tuning Result:" , '\n' , grid_search_cv.best_params_)

Fine_Tuning_Model_By_Grid_SearchCV(SupportVector_Machine_Model, X_Data_For_Train, Y_Data_For_Train)
```
